```python
from sklearn.model_selection import ParameterGrid
import pickle
import pandas as pd
import os

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

# Hyperparameters

```python
# Specifiy Parameters
params = {
    "lr": [1e-5],
    "epochs": [100],
    "batch_size": [16],
    "lookback": [20],
    "num_samples": [1, 2, 4, 8, 16],
    "paired_sampling": [True],
    "eff_lambda": [0.],
    "normalization": [None],
}

# Create Experiments as All Combinations
exp_list = list(ParameterGrid(params))

# Save Arguments To File
if not os.path.isdir('fastshap'):
    os.makedirs('fastshap')
with open('fastshap/arg_file.pkl', 'wb') as arg_file:
    pickle.dump(exp_list, arg_file)
```
